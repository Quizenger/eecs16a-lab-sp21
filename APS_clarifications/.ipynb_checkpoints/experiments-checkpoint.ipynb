# <span style="color:blue">Intro</span>

This deep dive into the backend of the APS lab system started during the testing of the virtual lab version of the APS 2 table experiment. If the receiver is placed too close to a beacon (speaker), there are large errors. However, if placed suffiiciently far away from all beacons, it works pretty well. 

1. Modulation and Demodulation Setup (Is this quadrature mod-demod as described [here](http://web.mit.edu/6.02/www/s2012/handouts/14.pdf)?)
1. Error calculation metric uses time delays v. difference between actual and calculated position vectors. Significant difference with the former performing better.
1. Improving (infinitely periodic linear) cross correlation plotting visuals (use windowing)
1. THOSE 3 Qs!

a. The beacons play on a loop at a fixed rate. However, we do not know when in the loop we started recording the beacon signals on the receiver. If we record for exactly the number of samples in one period of a beacon signal, we may have unwittingly started recording after some portion of the beacon signal has already been transmitted. -> **Non issue; periodic signal**

b. You may notice that the magnitude of the peaks in your cross-correlation output is dependent on the volume of the speakers. How might this affect the reliability of the positioning system? -> **Normalized cross-correlation?**

c. Like any real world system, the output will be affected by noise -> **Averaging**

### System Overview

Binary Signal {-1, 1} -> Shifted Binary Signal (by corresponding delay) -> Modulated by Multiplication with Carrier Signal (confusing which signal it is exactly, but a sinusoid) -> Demodulated with both sine (imag) and cosine (real) bases -> Passed through LPF -> Returned as sum of real and j*imag components

Makes sense, multiplication with rect function in frequency domain is equivalent to interpolation by convolution with sinc function in time domain.

```python
carrier_sig = Lab.beaconList[0].carrierSingal
carrier_sig_fft = np.fft.fft(carrier_sig)
carrier_len = Lab.beaconList[0].signalLength



print("Carrier Sig FT has imaginary component? {}".format(not(np.isreal(carrier_sig_fft).all())))
print("Carrier Sig FT has real component? {}".format((np.isreal(carrier_sig_fft).any())))
fig, axs = plt.subplots(3, figsize=[50, 50])
axs[0].plot(np.arange(0, 100), carrier_sig[:100])
axs[1].plot(np.arange(carrier_sig_fft.size), np.real(carrier_sig_fft))
axs[2].plot(np.arange(carrier_sig_fft.size), np.imag(carrier_sig_fft))
plt.show()
```

```python
print("Max value for Carrier Sig FT real part = {}".format(np.max(carrier_sig_fft)))
print("Max index for Carrier Sig FT real part = {}".format(np.argmax(carrier_sig_fft)))
print("Carrier Sig real part frequency = {} (added 1 to account for 0-index)".format(carrier_sig_fft.size//2 - np.argmax(carrier_sig_fft) + 1))
print("Carrier Sig real part = 54.25*cos(2330t)")
print()
print("Max value for Carrier Sig FT imag part = {}".format(np.max(np.imag(carrier_sig_fft))))
print("Max index for Carrier Sig FT imag part = {}".format(np.argmax(np.imag(carrier_sig_fft))))
print("Carrier Sig imag part frequency = -{} (added 1 to account for 0-index)".format(carrier_sig_fft.size//2 - np.argmax(np.imag(carrier_sig_fft)) + 1))
print("Carrier Sig imag part = 5090.66*sin(-2330t)")
print()
print("Carrier Freq according to code = {}".format(Lab.carrierFreq))
```

```python
binary_sig = Lab.beaconList[0].binarySignal
binary_sig_fft = np.fft.fft(binary_sig)

print("Binary Sig has imaginary component? {}".format(not(np.isreal(binary_sig).all())))
print("Carrier Sig FT has all real components? {}".format((np.isreal(binary_sig).all())))

fig, axs = plt.subplots(3, figsize=[50, 50])
axs[0].plot(np.arange(0, 10220), binary_sig)
axs[1].plot(np.arange(0, 10220), np.real(binary_sig_fft))
axs[2].plot(np.arange(0, 10220), np.imag(binary_sig_fft))
plt.show()
```
